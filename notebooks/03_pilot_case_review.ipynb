# Pilot Case Review

Inspection notebook for the pilot coding subset. This notebook reads the pilot CSV and summarizes coverage and missingness without modifying any data.

```python
from pathlib import Path

import pandas as pd

pd.set_option("display.max_rows", 200)
pd.set_option("display.max_columns", 200)
```

```python
project_root = Path.cwd()
pilot_path = project_root / "data" / "interim" / "pilot_coding_sheet.csv"

if not pilot_path.exists():
    project_root = Path.cwd().parent
    pilot_path = project_root / "data" / "interim" / "pilot_coding_sheet.csv"

if not pilot_path.exists():
    raise FileNotFoundError(f"Pilot dataset not found at {pilot_path}")

pilot_path
```

```python
df = pd.read_csv(pilot_path)
df.head()
```

## Dataset Shape

```python
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

## Row Counts By Case

```python
row_counts = (
    df["case_name"]
    .value_counts()
    .rename_axis("case_name")
    .reset_index(name="row_count")
)

row_counts
```

## Time Windows For Each Case

```python
time_windows = (
    df[["case_name", "case_id", "period_start", "period_end", "time_window_years"]]
    .sort_values(["case_name", "period_start", "period_end"])
    .reset_index(drop=True)
)

time_windows
```

## Fields That Are Still Empty

```python
missing_summary = (
    pd.DataFrame({
        "missing_count": df.isna().sum(),
        "missing_pct": df.isna().mean().mul(100),
    })
    .sort_values(["missing_count", "missing_pct"], ascending=False)
)

missing_summary[missing_summary["missing_count"] > 0]
```

```python
fully_empty_fields = missing_summary[missing_summary["missing_count"] == len(df)]
fully_empty_fields
```
