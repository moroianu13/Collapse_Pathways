# Dataset Overview

Starter notebook for inspecting the expanded collapse dataset without modifying source data.

```python
from pathlib import Path

import pandas as pd

project_root = Path.cwd()
data_path = project_root / "data" / "raw" / "collapse_dataset_expanded.csv"

if not data_path.exists():
    project_root = Path.cwd().parent
    data_path = project_root / "data" / "raw" / "collapse_dataset_expanded.csv"

data_path
```

```python
if not data_path.exists():
	alternative_path = next(
		project_root.rglob("collapse_dataset_expanded.csv"), None
	)
	if alternative_path is not None:
		data_path = alternative_path
	else:
		raise FileNotFoundError(
			f"Could not find collapse_dataset_expanded.csv under {project_root}"
		)

print(f"Reading data from: {data_path}")
df = pd.read_csv(data_path)

print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
df.head()
```

```python
df.columns.to_list()
```

```python
missing_summary = (
    df.isna()
    .sum()
    .rename("missing_values")
    .to_frame()
    .assign(missing_pct=lambda frame: frame["missing_values"] / len(df) * 100)
    .sort_values("missing_values", ascending=False)
)

missing_summary
```

```python
duplicate_case_ids = df[df.duplicated(subset="case_id", keep=False)].sort_values("case_id")
duplicate_case_ids
```
