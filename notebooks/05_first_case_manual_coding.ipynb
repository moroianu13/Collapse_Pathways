# First Case Manual Coding

Helper notebook for reviewing the Soviet Union rows in the pilot coding sheet before manual source entry. This notebook only reads and displays data.

```python
from pathlib import Path
import sys

import pandas as pd

pd.set_option("display.max_rows", 200)
pd.set_option("display.max_columns", 200)
```

```python
project_root = Path.cwd()
pilot_path = project_root / "data" / "interim" / "pilot_coding_sheet.csv"

if not pilot_path.exists():
    project_root = Path.cwd().parent
    pilot_path = project_root / "data" / "interim" / "pilot_coding_sheet.csv"

if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from src.dataset_utils import get_factor_columns

if not pilot_path.exists():
    raise FileNotFoundError(f"Pilot dataset not found at {pilot_path}")

pilot_path
```

```python
df = pd.read_csv(pilot_path)
soviet_rows = df[df["case_name"] == "Soviet Union"].copy()

print(f"Rows for Soviet Union: {len(soviet_rows)}")
soviet_rows.head()
```

## Relevant Rows

```python
metadata_columns = [
    "case_id",
    "case_name",
    "case_type",
    "region",
    "period_start",
    "period_end",
    "time_window_years",
    "phase_label",
    "collapse_outcome",
    "collapse_within_next_window",
    "data_confidence",
    "source_count",
    "notes",
]

soviet_rows[metadata_columns].sort_values(["period_start", "period_end"]).reset_index(drop=True)
```

## Factor Columns

```python
factor_columns = get_factor_columns()
factor_columns
```

```python
soviet_rows[["case_id", "period_start", "period_end"] + factor_columns].sort_values(["period_start", "period_end"]).reset_index(drop=True)
```

```python
factor_matrix = (
    soviet_rows[["case_id", "period_start", "period_end"] + factor_columns]
    .sort_values(["period_start", "period_end"])
    .set_index(["case_id", "period_start", "period_end"])
    .transpose()
)

factor_matrix
```
